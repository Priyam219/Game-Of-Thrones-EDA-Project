```python
#A BARPLOT FOR ATTACKER KING AND ARMY SIZE
plt.figure(figsize = (10,5))
sea.set(style = "whitegrid")
sea.barplot (x = 'attacker_king' , y = 'attacker_size' , data = got)
```
